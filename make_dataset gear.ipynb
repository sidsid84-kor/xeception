```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

def crop_image_opencv(image_path, text_file_path, save_path, saving_name):
    # 이미지 로드
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    height, width = img.shape[:2]
    try:
        # 텍스트 파일에서 중점 좌표와 사각형 크기 읽기
        with open(text_file_path, 'r') as file:
            data = file.readline().split()
            _, center_x, center_y, rect_width, rect_height = map(float, data)

            # 실제 픽셀 좌표로 변환
            center_x *= width
            center_y *= height
            rect_width *= width
            rect_height *= height

            # 여유분 추가 (각 방향으로 5%씩)
            extra_width = rect_width * 0.02
            extra_height = rect_height * 0.02
            rect_width += 2 * extra_width
            rect_height += 2 * extra_height

            # 잘라낼 영역 계산
            left = int(max(center_x - rect_width / 2, 0))
            top = int(max(center_y - rect_height / 2, 0))
            right = int(min(center_x + rect_width / 2, width))
            bottom = int(min(center_y + rect_height / 2, height))

            # 이미지 자르기
            cropped_img = img[top:bottom, left:right]

            # 잘라낸 이미지 저장
            resized_img = cv2.resize(cropped_img, (512, 512))

            # # 임계처리 
            # maxValue = 255
            # sub_value = 5
            # block_size = 19
            # thres45 = cv2.adaptiveThreshold(resized_img, maxValue, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, block_size, sub_value)

            # # 꼭지점 찾기
            # kernel_size=2
            # kernel = np.ones((kernel_size,kernel_size), np.uint8)

            # # Opening 연산 적용
            # opened_img = cv2.morphologyEx(thres45, cv2.MORPH_OPEN, kernel)

            # # 노이즈 제거를 위한 미디언 블러
            # blur=3
            # blurred = cv2.medianBlur(opened_img, blur)

            cv2.imwrite(save_path+saving_name+'.jpg', resized_img)

    except Exception as e:
        # print(e)
        pass
```
