```python
import pandas as pd
import cv2
import os
import shutil
import random
```

```python
data_path = r'E:\middle_broken'
good_path = data_path + '/good/'
broken_path = data_path + '/middle_broken/'
target_path = data_path + '/middle_data/'

good_list = os.listdir(good_path)
print(len(good_list))
broken_list = os.listdir(broken_path)
print(len(broken_list))
```

```python
df_empty = pd.DataFrame(columns=['id', 'good', 'defect'])
```

```python
sampled_files = random.sample(good_list, 7000)

for file_name in sampled_files:
    shutil.copy(good_path + file_name, target_path + file_name)
    new_row = pd.DataFrame({'id': file_name, 'good': 1, 'defect': 0}, index=[0])
    df_empty = pd.concat([df_empty, new_row], ignore_index=True)
df_empty
```

```python
select_num = 1
for file_name in broken_list:

    image = cv2.imread(broken_path + file_name)
    
    if select_num % 2:
    ##플립 위아래
        ud_image = cv2.flip(image,0)
        cv2.imwrite(target_path +'ud_'+ file_name, ud_image)
        new_row = pd.DataFrame({'id': 'ud_'+ file_name, 'good': 0, 'defect': 1}, index=[0])
        df_empty = pd.concat([df_empty, new_row], ignore_index=True)
    else:
        ##플립 좌우
        lr_image = cv2.flip(image,1)
        cv2.imwrite(target_path +'lr_'+ file_name, lr_image)
        new_row = pd.DataFrame({'id': 'lr_'+ file_name, 'good': 0, 'defect': 1}, index=[0])
        df_empty = pd.concat([df_empty, new_row], ignore_index=True)

    shutil.copy(broken_path + file_name, target_path + file_name)
    new_row = pd.DataFrame({'id': file_name, 'good': 0, 'defect': 1}, index=[0])
    df_empty = pd.concat([df_empty, new_row], ignore_index=True)

    select_num += 1

df_empty
```

```python
df_empty.to_csv('csvs/middle241106.csv', index=False)
```
