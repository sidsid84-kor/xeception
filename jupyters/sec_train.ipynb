```python
good_ssslist = os.listdir(r"E:\lenseImage\20250218adddata\good")
data = {'filename' : good_ssslist}
goodoridf = pd.DataFrame(data)
```

```python
# 날짜를 추출하는 함수
def extract_date(filename):
    return filename.split('_')[1]

# 날짜 열 추가
goodoridf['date'] = goodoridf['filename'].apply(extract_date)

# 날짜 열을 datetime 형식으로 변환
goodoridf['date'] = pd.to_datetime(goodoridf['date'], format='%Y%m%d%H%M%S')

# 날짜로 정렬
df_sorted = goodoridf.sort_values('date', ascending=False)
df_sorted
```

```python
# good_df = df_sorted[:10000]
good_df = pd.DataFrame(columns=new_sec_binary_df.columns)
good_df['id'] = df_sorted['filename'][:10000]
good_df['good'] = 1
good_df['defect'] = 0
```

```python
good_df
```

```python
defect_list = os.listdir(r'E:\lenseImage\20250218adddata\defect')
defect_data = {'id':defect_list}
defect_df = pd.DataFrame(data=defect_data)
defect_df['good'] = 0
defect_df['defect'] = 1
```

```python
defect_df
```

```python
add_df = pd.concat([good_df, defect_df], ignore_index=True)
add_df.duplicated('id').sum()
print(len(add_df))
add_ddf = add_df.drop_duplicates('id', keep=False)
add_ddf
```
