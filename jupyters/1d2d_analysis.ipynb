```python
import pandas as pd
import os
import tqdm
import shutil
```

```python
directory_path = "e:/polar_data/"
df_2d = pd.read_csv( directory_path + '2D.csv')
df_1d = pd.read_csv( directory_path + '1D.csv')
```

```python
df_2d
```

```python
def check_files(dire,fname):
    file_path = os.path.join(dire, fname)
    return os.path.exists(file_path)

not_exist_list = []
for rows in tqdm.tqdm(df_2d.itertuples(), total=len(df_2d)):
    if check_files(directory_path + '2D', rows.id):
        continue
    else:
        not_exist_list.append(rows.id)
print('2D 존재하지않는 파일수 : ', not_exist_list)

not_exist_list = []
for rows in tqdm.tqdm(df_1d.itertuples(), total=len(df_1d)):
    if check_files(directory_path + '1D', rows.id):
        continue
    else:
        not_exist_list.append(rows.id)
print('1D 존재하지않는 파일수 : ', not_exist_list)
```

```python
df_2d_good = df_2d[df_2d['good'] == 1]
df_2d_bubble = df_2d[df_2d['bubble'] == 1]
df_2d_edge = df_2d[df_2d['edge'] == 1]
df_2d_wrong = df_2d[(df_2d['good'] == 1) & ((df_2d['edge'] == 1) | (df_2d['bubble'] == 1))]
df_2d_notcheck = df_2d[(df_2d['good'] == 0) & (df_2d['edge'] == 0) & (df_2d['bubble'] == 0)]

print('총개수', len(df_2d))
print('good', len(df_2d_good))
print('bubble', len(df_2d_bubble))
print('edge', len(df_2d_edge))
print('잘못체크됨', len(df_2d_wrong))
print('체크안됨', len(df_2d_notcheck))
```

```python
for rows in tqdm.tqdm(df_2d_notcheck.itertuples(), total=len(df_2d_notcheck)):
    if check_files(directory_path + '2D', rows.id):
        shutil.copy(directory_path + '2D/' + rows.id, directory_path + 'check/' + rows.id)
    else:
        print("???")
```

```python
df_1d_good = df_1d[df_1d['good'] == 1]
df_1d_bubble = df_1d[df_1d['bubble'] == 1]
df_1d_edge = df_1d[df_1d['edge'] == 1]
df_1d_wrong = df_1d[(df_1d['good'] == 1) & ((df_1d['edge'] == 1) | (df_1d['bubble'] == 1))]
df_1d_notcheck = df_1d[(df_1d['good'] == 0) & (df_1d['edge'] == 0) & (df_1d['bubble'] == 0)]


print('good', len(df_1d_good))
print('bubble', len(df_1d_bubble))
print('edge', len(df_1d_edge))
print('잘못체크됨', len(df_1d_wrong))
print('체크안됨', len(df_1d_notcheck))
```

```python
for rows in tqdm.tqdm(df_1d_wrong.itertuples(), total=len(df_1d_wrong)):
    if check_files(directory_path + '1D', rows.id):
        shutil.copy(directory_path + '1D/' + rows.id, directory_path + 'check/' + rows.id)
    else:
        print("???")
```
